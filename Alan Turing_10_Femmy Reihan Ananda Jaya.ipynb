```python
import sqlite3
import pandas as pd
```

# Connect sqlite3 to python

```python
connection = sqlite3.connect('data.sqlite')
```

# Operasi Insert

```python
result = connection.cursor()
command = "INSERT INTO AGENTS VALUES ('A013', 'Achmad Malik', 'Indonesia', '0.27', '021-14045', '')"
result.execute(command)
```

# Operasi Select

```python
command = "select * from agents"
result.execute(command)
results = result.fetchall()
cols = [column[0] for column in result.description]
results_data= pd.DataFrame.from_records(results, columns = cols)

results_data
```

# 1. Tampilkan data customer yang mempunyai opening amount diatas 7000

```python
command = "select * from customer WHERE OPENING_AMT > '7000';"
result.execute(command)
results = result.fetchall()
cols = [column[0] for column in result.description]
results_data= pd.DataFrame.from_records(results, columns = cols)

results_data
```

# 2. Tampilkan informasi order yang dilakukan melalui agensi dengan code A003



```python
command = "SELECT * FROM orders WHERE AGENT_CODE IN ('A003');"
result.execute(command)
results = result.fetchall()
cols = [column[0] for column in result.description]
results_data= pd.DataFrame.from_records(results, columns = cols)

results_data
```

# 3. Tampilkan data customer yang tinggal di New York, Bangalore, Mumbai, dan Torento

```python
command = "SELECT * FROM customer WHERE CUST_CITY IN ('New York', 'banglore','Mumbai','Torento');"
result.execute(command)
results = result.fetchall()
cols = [column[0] for column in result.description]
results_data= pd.DataFrame.from_records(results, columns = cols)

results_data
```

# 4. Tampilkan informasi customer yang mempunyai payment amount diatas 7000 dan urutkan berdasarkan opening amountnya dari yang terbesar



```python
command = "select * from customer WHERE PAYMENT_AMT > '7000';"
result.execute(command)
results = result.fetchall()
cols = [column[0] for column in result.description]
results_data= pd.DataFrame.from_records(results, columns = cols)

results_data
```

# 5. Tampilkan data agensi yang tidak bekerja di London dan memiliki comission lebih dari 0.12

```python
result = connection.cursor()
command = "select * from agents WHERE NOT WORKING_AREA = 'London' AND COMMISSION >= '0.12'"
result.execute(command)
results = result.fetchall()
cols = [column[0] for column in result.description]
results_data= pd.DataFrame.from_records(results, columns = cols)

results_data
```
